Task 3 – Create SaaS Customer Class (OOP) 

```python
class SaaSCustomer:
   
    def __init__(self, customer_id, name, signup_date, plan, monthly_fee, usage_hours):
        self.customer_id = customer_id
        self.name = name
        self.signup_date = signup_date
        self.plan = plan
        self.monthly_fee = monthly_fee
        self.usage_hours = usage_hours

    def validate_data(self):
        self.monthly_fee=float(self.monthly_fee)
        
        if not self.name or str(self.name).strip().lower() in ['none','null','nan']:
            self.name='Unknown Customer'
        
        self.plan= str(self.plan).strip().capitalize() if self.plan else 'unknown'

        hours=str(self.usage_hours).lower

        if hours in ['null','none','nan']:
            self.usage_hours=0
        else:
            self.usage_hours=int(self.usage_hours)

        print('Monthly fees',self.monthly_fee)
        print('Name',self.name)
        print('UsageHours',self.usage_hours)
        print('plan',self.plan)
    
        
    def display_customer(self):
        
        print(f'Customer Name={self.name}')






```

Task 4 – Data Cleaning Pipeline 

```python
import datetime
```

```python
from datetime import datetime

def handle_missing_values(data):
   
    cleaned_names = []
    for name in data["name"]:
        if name is None or (isinstance(name, str) and name.strip() == ""):
            cleaned_names.append("Unknown Customer")
        else:
            cleaned_names.append(name)
    data["name"] = cleaned_names

    cleaned_hours = []
    for hours in data["usage_hours"]:
        if hours is None or str(hours).strip().upper() in ["NULL", ""]:
            cleaned_hours.append(0)
        else:
            try:
               
                cleaned_hours.append(int(hours))
            except ValueError:
                cleaned_hours.append(0)
    data["usage_hours"] = cleaned_hours

   
    cleaned_fees = []
    for fee in data["monthly_fee"]:
        if fee is None:
            cleaned_fees.append(None)
        else:
            try:
                
                float_value = float(fee)
                cleaned_fees.append(str(float_value)) 
            except ValueError:
                cleaned_fees.append(None)
    data["monthly_fee"] = cleaned_fees

    return data


def remove_duplicates(data):
    seen_ids = set()
    unique_indices = []
  
    for index, name in enumerate(data["name"]):
        if name not in seen_ids:
            unique_indices.append(index)
            seen_ids.add(name)
            
    cleaned_data = {}
    for key, values in data.items():
        cleaned_data[key] = [values[i] for i in unique_indices]
        
    return cleaned_data


def standardize_dates(data):
    date_formats = ["%Y-%m-%d", "%Y/%m/%d", "%d-%m-%Y", "%Y%m%d"]
    standardized_list = []
    
    for date_str in data["signup_date"]:
        parsed_date = None
        for fmt in date_formats:
            try:
                parsed_date = datetime.strptime(str(date_str), fmt)
                break
            except ValueError:
                continue
                
        if parsed_date:
            standardized_list.append(parsed_date.strftime("%Y-%m-%d"))
        else:
            standardized_list.append(date_str)
            
    data["signup_date"] = standardized_list
    return data


def data_cleaning_pipeline(raw_data):
   
    data = remove_duplicates(raw_data)
    data = handle_missing_values(data)
    data = standardize_dates(data)
    return data
```

```python
Cleaned_saas_data=data_cleaning_pipeline(saas_data)
Cleaned_saas_data
```
